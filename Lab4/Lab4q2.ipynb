```python
import random
import matplotlib.pyplot as plt

#  Utility functions (from scratch) 
def compute_mean(data):
    return sum(data) / len(data)

def compute_variance(data):
    mean = compute_mean(data)
    return sum((x - mean) ** 2 for x in data) / len(data)

# Binomial Simulation
def simulate_binomial(trials, prob_success, experiments):
    """Simulate binomial distribution by doing experiments of 'trials' with success probability."""
    results = []
    for _ in range(experiments):
        success_count = sum(1 for _ in range(trials) if random.random() < prob_success)
        results.append(success_count)
    return results

def plot_histogram(data, title, bins=20):
    plt.hist(data, bins=bins, edgecolor='black', alpha=0.7)
    plt.title(title)
    plt.xlabel("Number of Successes")
    plt.ylabel("Frequency")
    plt.show()

# Main Binomial Problem 
trials = 100
experiments = 10000
success_rates = [0.90, 0.95, 0.97, 0.99]  # test with different probabilities

for p in success_rates:
    data = simulate_binomial(trials, p, experiments)
    mean_val = compute_mean(data)
    var_val = compute_variance(data)
    print(f"p={p}, Mean={mean_val:.2f}, Variance={var_val:.2f}")
    plot_histogram(data, f"Binomial Distribution (p={p}, trials={trials})")
```
